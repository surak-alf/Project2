# Statistical Summary

```python
df.describe()
```

```python
min_date = min(df['Timestamp'])
max_date = max(df['Timestamp'])

print("Earliest date:", min_date)
print("Latest date:", max_date)
```

```python
start_date = pd.to_datetime('2021-08-09')
end_date = pd.to_datetime('2022-08-08')

df['half_year'] = pd.cut(df['Timestamp'], bins=[start_date, pd.Timestamp('2022-02-01'), end_date], labels=['First Half', 'Second Half'])
```

```python
grouped_df = df.groupby('half_year')
```

```python
for half_year, group_df in grouped_df:
    plt.figure(figsize=(12, 6))
    group_df.plot(x='Timestamp', y=group_df.columns[1:], figsize=(12, 6))
    plt.title(f'Data for {half_year}')
    plt.xlabel('Timestamp')
    plt.ylabel('Value')
    plt.xticks(rotation=45)
    plt.legend(loc='upper right')
    plt.show()
```

```python
df['quarter'] = df['Timestamp'].dt.quarter
# Group the data by quarter
grouped_df = df.groupby('quarter')
```

```python
# Plot the data for each quarter
for quarter, group_df in grouped_df:
    plt.figure(figsize=(12, 6))
    group_df.plot(x='Timestamp', y=group_df.columns[1:], figsize=(12, 6))
    plt.title(f'Data for Quarter {quarter}')
    plt.xlabel('Timestamp')
    plt.ylabel('Value')
    plt.xticks(rotation=45)
    plt.legend(loc='upper right')
    plt.show()
```
